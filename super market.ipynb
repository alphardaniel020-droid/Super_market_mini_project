```python
#oulier
Q1=np.percentile(df['Profit'],25,method='midpoint')
Q2=np.percentile(df['Profit'],50,method='midpoint')
Q3=np.percentile(df['Profit'],75,method='midpoint')
print(Q1)
print(Q2)
print(Q3)
```

```python
df['Profit'].median()
```

```python
IQR=Q3-Q1
print(IQR)
```

```python
upperlimit=Q3+1.5*IQR
lowerlimit=Q1-1.5*IQR
print(upperlimit)
print(lowerlimit)
```

```python
outlier=[]
for i in df['Profit']:
    if((i > upperlimit),(i <lowerlimit)):
        outlier.append(i)
outlier
```

```python
colm=['Postal Code','Sales','Quantity','Discount','Profit']
df_cleaned=df.copy()

for col in colm:
    Q1 =df[col].quantile(0.25)
    Q3=df[col].quantile(0.75)
    IQR=Q3-Q1
    df_cleaned = df_cleaned[
        (df_cleaned[col] >= Q1-1.5*IQR)&
        (df_cleaned[col] <= Q3+ 1.5*IQR)
    ]
print("Before:" , df.shape)
print("After:", df_cleaned.shape)
```
